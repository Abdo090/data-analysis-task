```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("Iris.csv")
```

```python
for column in data.columns:
    plt.hist(data[column], edgecolor='black')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.grid(True)
    plt.show()
```
